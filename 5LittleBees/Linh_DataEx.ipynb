```python
import matplotlib.pyplot as plt
```

```python
fixed_acid = red_df['fixed acidity']
volatile_acid = red_df['volatile acidity']
citric_acid = red_df['citric acid']
res_sugar = red_df['residual sugar']
chloride = red_df['chlorides']
free_so2 = red_df['free sulfur dioxide']
tot_so2 = red_df['total sulfur dioxide']
density = red_df['density']
pH = red_df['pH']
sulphates = red_df['sulphates']
alcohol = red_df['alcohol'] 
quality = red_df['quality']
```

```python
count_quality = np.array(red_df['quality'].value_counts())
list_quality = red_df['quality'].value_counts().index
quality_percent = []
for i in range(len(count_quality)):
    quality_percent.append(count_quality[i]/1599*100)
```

```python
# Draw graph to visualize each feature

from random import random
from turtle import color, width

plot_objects = plt.subplots(nrows=6, ncols=2, figsize=(25, 40))
fig, ((ax1, ax2), (ax3, ax4), (ax5, ax6), (ax7,ax8), (ax9,ax10), (ax11, ax12)) = plot_objects
ax1.hist(fixed_acid, bins=100, alpha=0.5, histtype= 'bar', color='#f2b852', edgecolor='#f2b852'), ax1.title.set_text('Fixed acid')
ax2.hist(volatile_acid, bins=100, alpha=0.5, histtype= 'bar', color='#63d128', edgecolor='#63d128'),  ax2.title.set_text('Volatile acid')
ax3.hist(citric_acid, bins=100, alpha=0.5, histtype= 'bar', color='#aeebe5', edgecolor='#aeebe5'),  ax3.title.set_text('Citric acid')  
ax4.hist(res_sugar, bins=100, alpha=0.5, histtype= 'bar', color='#d6badb', edgecolor='#d6badb'),  ax4.title.set_text('Residual sugar')
ax5.hist(chloride, bins=100, alpha=0.5, histtype= 'bar', color='#e388a9', edgecolor='#e388a9'),  ax5.title.set_text('Chlorides')
ax6.hist(free_so2, bins=100, alpha=0.5, histtype= 'bar', color='#544fdb', edgecolor='#544fdb'),  ax6.title.set_text('Free sulfur dioxide')  
ax7.hist(tot_so2, bins=100, alpha=0.5, histtype= 'bar', color='#e64e4c', edgecolor='#e64e4c'),  ax7.title.set_text('Total sulfur dioxide')
ax8.hist(density, bins=100, alpha=0.5, histtype= 'bar', color='#f27e2c', edgecolor='#f27e2c'),  ax8.title.set_text('Density')
ax9.hist(pH, bins=100, alpha=0.5, histtype= 'bar', color='#325e26', edgecolor='#325e26'),  ax9.title.set_text('pH')
ax10.hist(sulphates, bins=100, alpha=0.5, histtype= 'bar', color='#488ce0', edgecolor='#488ce0'),  ax10.title.set_text('Sulphates')
ax11.hist(alcohol, bins=100, alpha=0.5, histtype= 'bar', color='#f5f247', edgecolor='black'),  ax11.title.set_text('Alcohol')
ax12.pie(quality_percent, labels=list_quality),  ax12.title.set_text('Quality')
```
