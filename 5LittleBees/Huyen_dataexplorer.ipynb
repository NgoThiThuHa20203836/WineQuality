```python
#Correlation
corr=white_wine.corr()
corr
```

```python
#Corr matrix in coolwarm
corr.style.background_gradient(cmap='coolwarm')
```

```python
#plot corr matrix
fig = plt.figure(figsize=(7,7))
plt.matshow(corr, cmap='RdBu', fignum=fig.number)
plt.xticks(range(len(corr.columns)), corr.columns, rotation='vertical');
plt.yticks(range(len(corr.columns)), corr.columns);
```

```python
#plot corr heatmap
sns.heatmap(white_wine.corr(),annot=True)
```

```python
#plot compare with fixed acidity
white_wine.corr().plot(kind='scatter', x='fixed acidity', y='volatile acidity', figsize=(6,6))
white_wine.corr().plot(kind='scatter', x='fixed acidity', y='citric acid', figsize=(6,6))
white_wine.corr().plot(kind='scatter', x='fixed acidity', y='residual sugar', figsize=(6,6))
white_wine.corr().plot(kind='scatter', x='fixed acidity', y='chlorides', figsize=(6,6))
white_wine.corr().plot(kind='scatter', x='fixed acidity', y='free sulfur dioxide', figsize=(6,6))
white_wine.corr().plot(kind='scatter', x='fixed acidity', y='total sulfur dioxide', figsize=(6,6))
white_wine.corr().plot(kind='scatter', x='fixed acidity', y='density', figsize=(6,6))
white_wine.corr().plot(kind='scatter', x='fixed acidity', y='pH', figsize=(6,6))
white_wine.corr().plot(kind='scatter', x='fixed acidity', y='sulphates', figsize=(6,6))
white_wine.corr().plot(kind='scatter', x='fixed acidity', y='alcohol', figsize=(6,6))
white_wine.corr().plot(kind='scatter', x='fixed acidity', y='quality', figsize=(6,6))
```
